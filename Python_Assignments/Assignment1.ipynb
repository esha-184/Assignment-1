## Assignment Question 1
Design a simple calculator using nested conditional statements.


```python
num1=float(input("Enter the first number:"))
num2=float(input("Enter the second number:"))
operation=input("Enter the operation(+,-,*,/):")

if operation == '+':
   result = num1 + num2
elif operation == '-':
   result = num1 - num2
elif operation == '*':
    result = num1 * num2
elif operation == '/':
      if num2!=0:
          result = num1 / num2
      else :
          result = ("Error! Division by zero")
else :
    result = "Invalid operation"
print("Result:",result)
```

## Assignment Question 2
Determine the ticket price based on age and whether the person is a student.

```python
age=int(input("Enter the age:"))
is_student=input("Are you a student?(yes/no):")

if age <= 7:
  price = "Free"
elif age <= 10:
  price = "Free"
elif age <= 17:
  if is_student == "yes":
    price = "10$"
  else:
    price = "12$"
elif age <= 20:
  if is_student == "yes":
    price = "15$"
  else:
    price = "18$"
elif age <= 22:
  if is_student == "yes":
    price = "20$"
  else:
    price = "25$"
else:
  price = "30$"

print("Ticket Price:", price)
```

## Assignment Question 3
Determine whether the year is a leap year or not.


```python
year=int(input("Enter the year:"))

if year%4==0:
    if year%100==0:
        if year%400==0:
            print(year,"is a leap year")
        else:
            print(year,"is not a leap year")

    else:
         print(year,"is a leap year")
else:
    print(year,"is not a leap year")
```

## Assignment Question 5
Hospital emergency priority system.


```python
symptoms = input("Enter symptom level (severe/moderate/mild): ").lower()

if symptoms == "severe":

    age = int(input("Enter patient's age: "))

    if age >= 65:
        print("Priority: Critical")
        print("Patient should be sent directly to the ICU.")
    else:
        print("Priority: High")
        print("Patient should be admitted to the Emergency Ward.")

elif symptoms == "moderate":

    temperature = float(input("Enter patient's temperature (°F): "))

    if temperature > 102:
        print("Patient should be admitted to the Emergency Ward.")
    else:
        print("Patient should be sent to the General Doctor.")

elif symptoms == "mild":

    age = int(input("Enter patient's age: "))

    if age < 12:
        print("Patient should be sent to the Pediatric Department.")
    else:
        print("Give the patient a regular appointment.")

else:
    print("Invalid symptom level entered.")


```
